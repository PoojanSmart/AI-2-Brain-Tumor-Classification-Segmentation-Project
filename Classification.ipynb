```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
```

```python
datagen = ImageDataGenerator(rescale = 1/255)
```

```python
traingen = datagen.flow_from_directory("classification/train",target_size=(512, 512), color_mode='grayscale', 
                            class_mode='categorical', batch_size=16, shuffle=True)
valgen = datagen.flow_from_directory("classification/validation",target_size=(512, 512), color_mode='grayscale', 
                            class_mode='categorical', batch_size=16, shuffle=True)
testgen = datagen.flow_from_directory("classification/test",target_size=(512, 512), color_mode='grayscale', 
                            class_mode='categorical', batch_size=16, shuffle=True)
```

```python
class_maps = dict([(value, key) for key, value in traingen.class_indices.items()])
```

```python
f,ax = plt.subplots(4,4, figsize=(12,12))
plt.tight_layout()
for i,axs in enumerate(ax.ravel()):
    axs.imshow(traingen[0][0][i])
    label = np.argmax(traingen[0][1][i])
    axs.set_title("Class:"+class_maps[label])
plt.show()
```
